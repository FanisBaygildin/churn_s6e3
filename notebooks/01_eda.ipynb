```python
def quick_eda(df, target=None):
    display(train.info())
    print()
    
    print("MISSING VALUES")
    miss = df.isna().sum().sort_values(ascending=False)
    miss = miss[miss > 0]
    if len(miss) == 0:
        print("No missing values")
    else:
        display(pd.DataFrame({
            "missing_cnt": miss,
            "missing_pct": (miss / len(df) * 100).round(2)
        }))
    print()    # пустая строка

    print("NUNIQUE")
    nunq = df.nunique(dropna=False).sort_values()
    display(nunq.to_frame("nunique"))
    print()

    print("NUMERICAL DESCRIBE")
    display(df.describe().T)
    print()

    print("CATEGORICAL DESCRIBE")
    cat_cols = df.select_dtypes(include=["object", "str", "category", "bool"]).columns
    if len(cat_cols) > 0:
        display(df[cat_cols].describe().T)
    else:
        print("No categorical columns")
    print()
    
    if target is not None and target in df.columns:
        print("TARGET DISTRIBUTION")
        display(df[target].value_counts(dropna=False).to_frame("count"))
        display((df[target].value_counts(normalize=True, dropna=False) * 100).round(2).to_frame("pct"))

    for col in cat_cols:
        vc = df[col].value_counts(normalize=True, dropna=False)
        
        values = [
            f"{k} ({v:.2%})"
            for k, v in vc.items()
        ]
        
        print(f"{col}: {values}")
```

```python
quick_eda(train, target="Churn")
```
